# Example of how to load and display results


```python
import numpy as np
import matplotlib.pyplot as plt
```

Load a file, it is named like "occupancy_info_e1gpc_5.0_0001_sl.npy", where 
- e1gpc refers to phase and polarisation
- 5.0 refers to xi value
- 0001 to generated bunch crossing
- sl to the used Simplified Simulation setup.

```python
data = np.load("/nfs/dust/luxe/user/spatarod/SimplifiedLUXE/e-laser/phase1/gpc/5.0/occupancy/" \
               "occupancy_info_e1gpc_5.0_0001_fl.npy", 
               allow_pickle=True)[()][0]
```

## Data structure

- single hit: each particle triggers exactly one pixel
- single hit net: net number of triggered pixels
- multi hit: particles may trigger more than one pixel
- multi hit net: net number of triggered pixels
- histogram single: num of hits per pixel, single pixel hit case
- histogram multi: num of hits per pixel, multi pixel hit case

```python
print(data.keys())
```

Within the valuees of the top level dictionary, everything is organized via adressing the planes, whose names are the keys of the nested dictionaries

For "sl", the simplified LUXE setup (18 chips reduced to 1 chip per layer, no gaps) the actual pixel size is made 8 times as large as in reality. This is done because the 2d histogram would be a 16384x512 matrix otherwise.

For the "fl", the full LUXE setup case, the pixel size is set as the one used in the actual LUXE experiment.

## Plotting the histograms

```python
data["histogram single"].keys()
```

```python
# Choose a plane
data_to_plot = data["histogram single"]["Histogram data of Plane 31"]
```

```python
plt.figure(figsize=(9,6),dpi=200)
plt.bar(np.arange(len(data_to_plot)), 
        data_to_plot,
        label=f"detector Plane 31\nN={int(data['single hit net']['Net single hits on Plane 31'])}")
plt.yscale("log")
plt.xlabel("hits/pixel", fontsize=18)
plt.ylabel("number of pixels", fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=18)
plt.grid(True, which="both", ls='--', linewidth=0.5)
```
