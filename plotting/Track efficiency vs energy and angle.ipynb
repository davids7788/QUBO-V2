# This notebook is a blueprint for plotting track efficiency/fake rate vs. particle energy/track angle

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Energy,  found tracks, efficiency, fake rate

data = {1500: [1, 1.0, 0.0], 
        2000: [30, 0.93, 0.07],
        2500: [119, 1.0, 0.0],
        3000: [233, 0.98, 0.02],
        3500: [299, 1.0, 0.0],
        4000: [315, 0.98, 0.02],
        4500: [336, 0.97, 0.03],
        5000: [225, 0.97, 0.03],
        5500: [199, 0.96, 0.04],
        6000: [157, 0.95, 0.05],
        6500: [94, 1.0, 0.0],
        7000: [59, 1.0, 0.0],
        7500: [33, 0.94, 0.06],
        8000: [10, 1.0, 0.0],
        8500: [3, 1.0, 0.0]}

data2 = {1500: [22, 1.0, 0.0],
         2000: [288, 0.97, 0.03],
         2500: [802, 0.97, 0.03],
         3000: [1292, 0.92, 0.08],
         3500: [1478, 0.89, 0.11],
         4000: [1502, 0.82, 0.18],
         4500: [1319, 0.79, 0.21],
         5000: [1102, 0.83, 0.17],
         5500: [866, 0.85, 0.15],
         6000: [605, 0.89, 0.11],
         6500: [446, 0.89, 0.11],
         7000: [233, 0.96, 0.04],
         7500: [127, 0.94, 0.06],
         8000: [74, 0.99, 0.01],
         8500: [31, 1.0, 0.0],
         9000: [12, 1.0, 0.0]}
```

```python
energy_xi4 = [key for key in data.keys()]
events_xi4 = [data[key][0] for key in data]
sum_events_xi4 = sum([data[key][0] for key in data])
efficiency_xi4 = [data[key][1] for key in data]
fake_rate_xi4 = [data[key][2] for key in data]

energy_xi5 = [key for key in data2.keys()]
events_xi5 = [data2[key][0] for key in data2]
sum_events_xi5 = sum([data2[key][0] for key in data2])
efficiency_xi5 = [data2[key][1] for key in data2]
fake_rate_xi5 = [data2[key][2] for key in data2]

fig, (ax_0, ax_1) = plt.subplots(2, figsize=(12,9), dpi=600)
ax_0_2 = ax_0.twinx()
# ax_1_2 = ax_1.twinx()

ax_0_2.plot(energy_xi4, events_xi4, label="found tracks", color="grey", 
            linestyle="--",linewidth=2.5, marker="^")
ax_0.plot(energy_xi4, np.array(efficiency_xi4) * 100, 
          label="efficiency", linestyle="--",color="b", linewidth=2.5, marker="o")
ax_0.plot(energy_xi4, np.array(fake_rate_xi4) * 100, 
          label="fake rate", linestyle="--",color="r", linewidth=2.5, marker="s")
ax_0.set_ylabel("efficiency / fake rate [%]", fontsize=14)
ax_0.legend(loc=6, fontsize=14)
ax_0.set_title(r"phase 0, $\xi$ =5, 2129 found tracks", loc="left", fontsize=14)
ax_0.tick_params(labelsize=14)
ax_0_2.set_ylabel("track count", fontsize=14)
ax_0_2.tick_params(labelsize=14)
ax_0.set_xlim(min(energy_xi4 + energy_xi5) - 750, max(energy_xi4 + energy_xi5) + 750)
ax_0_2.legend(loc=7, fontsize=14)


#ax_1_2.bar(energy_xi5, events_xi5, label="found tracks", color="grey", 
 #           linestyle="--",linewidth=1.5)
ax_1.bar(np.array(energy_xi5) 
```

```python
energy_xi4 = [key for key in data.keys()]
events_xi4 = [data[key][0] for key in data]
sum_events_xi4 = sum([data[key][0] for key in data])
efficiency_xi4 = [data[key][1] for key in data]
fake_rate_xi4 = [data[key][2] for key in data]

energy_xi5 = [key for key in data2.keys()]
events_xi5 = [data2[key][0] for key in data2]
sum_events_xi5 = sum([data2[key][0] for key in data2])
efficiency_xi5 = [data2[key][1] for key in data2]
fake_rate_xi5 = [data2[key][2] for key in data2]

fig, (ax_0, ax_1) = plt.subplots(2, figsize=(12,9), dpi=600)
ax_0_2 = ax_0.twinx()
ax_1_2 = ax_1.twinx()

ax_0_2.plot(energy_xi4, events_xi4, label="found tracks", color="grey", 
            linestyle="--",linewidth=2.5, marker="^")
ax_0.plot(energy_xi4, np.array(efficiency_xi4) * 100, 
          label="efficiency", linestyle="--",color="b", linewidth=2.5, marker="o")
ax_0.plot(energy_xi4, np.array(fake_rate_xi4) * 100, 
          label="fake rate", linestyle="--",color="r", linewidth=2.5, marker="s")
ax_0.set_ylabel("efficiency / fake rate [%]", fontsize=14)
ax_0.legend(loc=6, fontsize=14)
ax_0.set_title(r"phase 0, $\xi$ =5, 2129 found tracks", loc="left", fontsize=14)
ax_0.tick_params(labelsize=14)
ax_0_2.set_ylabel("track count", fontsize=14)
ax_0_2.tick_params(labelsize=14)
ax_0.set_xlim(min(energy_xi4 + energy_xi5) - 750, max(energy_xi4 + energy_xi5) + 750)
ax_0_2.legend(loc=7, fontsize=14)


ax_1_2.plot(energy_xi5, events_xi5, label="found tracks", color="grey", 
           linestyle="--",linewidth=1.5)
ax_1.bar(np.array(energy_xi5) - 100, np.array(efficiency_xi5) * 100, width=200,
          label="efficiency",color="b", align="edge")
ax_1.bar(np.array(energy_xi5) + 100, np.array(fake_rate_xi5) * 100, width=200, 
          label="fake rate", color="r", align="edge")
ax_1.set_xlabel("energy [MeV]", fontsize=14)
ax_1.set_ylabel("efficiency / fake rate [%]", fontsize=14)
ax_1_2.set_ylabel("track count", fontsize=14)
ax_1_2.tick_params(labelsize=14)
ax_1.set_title(r"phase 0, $\xi$ =5, 10216 found tracks", loc="left", fontsize=14)
ax_1.tick_params(labelsize=14)
ax_1.legend(loc=6, fontsize=14)
ax_1.set_xlim(min(energy_xi4 + energy_xi5) - 750, max(energy_xi4 + energy_xi5) + 750)
ax_1_2.legend(loc=7, fontsize=14)

plt.savefig("Efficiency_vs_Energy.pdf")
```
